```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Set the seaborn theme
sns.set()

# Load the 'mpg' dataset
df = sns.load_dataset('mpg')

# Display the first few rows of the dataset to understand its structure
df.head()

# Plot a scatter plot to check the relationship between 'mpg' and 'horsepower'
sns.relplot(x='mpg', y='horsepower', data=df)
plt.title('Scatter Plot of MPG vs Horsepower')
plt.show()

# Check if the two variables are related by inspecting the scatter plot
# Calculate the correlation coefficient
correlation = df['mpg'].corr(df['horsepower'])

# Print the result based on the correlation coefficient
if pd.notnull(correlation) and abs(correlation) > 0.5:
    print("Yes")
else:
    print("No")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Set the seaborn theme
sns.set()

# Load the 'mpg' dataset
df = sns.load_dataset('mpg')

# Display the first few rows of the dataset to understand its structure
df.head()

# Plot a scatter plot to check the relationship between 'mpg' and 'horsepower', colored by 'origin'
sns.relplot(x='mpg', y='horsepower', hue='origin', data=df)
plt.title('Scatter Plot of MPG vs Horsepower by Origin')
plt.show()

# Count the number of data points for each origin
origin_counts = df['origin'].value_counts()

# Print the origin with the highest count
majority_origin = origin_counts.idxmax()
print(majority_origin)
```
